### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Somente mais uma linha para teste


```python
import pyautogui as pg
import time

link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing"

# pyautogui.click -> clicar com o mouse
# pyautogui.write -> escrever com o teclado
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas

pg.PAUSE = 1  # tempo de espera entre as ações

# Passo a passo do desafio
# Passo 1: Entrar no sistema de empresa (simulado pelo link)
# abrir o chrome
pg.press("win")
# pg.hotkey("ctrl", "esc")  # abrir nova aba
pg.write("chrome")
pg.press("enter")

time.sleep(2)  # esperar o navegador abrir

pg.click(x=729, y=1797)

# entrar no link (https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing)
# apertar Enter e esperar a página carregar

pg.write(link)
time.sleep(2)  # esperar o navegador abrir
pg.press("enter")

# Passo 2: Navegar até encontrar a base de dados
# Passo 3: Baixar arquivo da base de dados
# Passo 4: Calcular os indicadores (faturamento e quantidade de produtos vendidos
# Passo 5: Enviar um e-mail para a diretoria com os indicadores

# Instalar as bibliotecas 
# pyautogui -> automações
# Pandas -> Dados
# openpyxl -> Manipular arquivos excel
# para instalar as bibliotecas use:
# pip install pyautogui pandas openpyxl

```
